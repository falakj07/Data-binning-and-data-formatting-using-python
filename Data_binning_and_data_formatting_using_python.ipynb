```python
import pandas as pd
import numpy as np

data = {
    'Product': ['Laptop', 'Mobile', 'Tablet', 'Headphones', 'Smartwatch', 'Camera'],
    'Price': [55000, 20000, 30000, 2000, 8000, 45000],
    'Units_Sold' : [25, 60, 35, 80, 50, 15]
}
```

```python
##data binning
bins= [0, 10000, 30000, 60000]
labels = ['Low', 'Medium', 'High']

df = pd.DataFrame(data)
df['Price_Category'] = pd.cut(df['Price'], bins=bins, labels=labels)
df
```

```python
#Binning Units Sold
bins2=[0,30,60,100]
labels2=['Low Sales', 'Medium Sales', 'High Sales']

df['Sales_Category'] = pd.cut(df['Units_Sold'],bins=bins2, labels=labels2)
df
```

```python
#Data Formatting
#Convert Data Type

print(df.dtypes)

df['Units Sold'] = df['Units_Sold'].astype(float)

df.dtypes
```

```python
#Convert text to uppercase

df['Product'] = df['Product'].str.upper()
df
```

```python
#Additional operations
#sort data
sorted_df = df.sort_values(by='Price')
sorted_df
```

```python
#Display unique values
print(df['Price_Category'].unique())
```

```python
#Dataset: Food Delivery Orders

data = {
    'Order_ID':[1,2,3,4,5,6,7,8],
    'Order_Value':[250,450,120,600,300,150,500,220],
    'Delivery_Time':[30,45,15,60,20,35,50,28],
    'Distance_km':[5,10,3,8,6,2,9,4]
}
df = pd.DataFrame(data)
df
```

```python
bins3=[0,20,40,60]
labels3=['Short','Medium','Long']
df['Delivery_Category'] = pd.cut(df['Delivery_Time'],bins=bins3, labels=labels3)
df
```

```python
bins4=[0,200,400,600]
labels4=['Affordable','Luxe','Gourmet']
df['MenuPrice_Category'] = pd.cut(df['Order_Value'],bins=bins4, labels=labels4)
df
```

```python
bins5 = [0,3,6,10]
labels5 = ['Close','Medium','Far']
df['Distance_Category'] = pd.cut(df['Distance_km'],bins=bins5,labels=labels5)
df
```

```python
#Data Formatting
#Convert Data Type

print(df.dtypes)

df['Distance_km'] = df['Distance_km'].astype(float)

df
```

```python
# Convert Text to Uppercase
df['Delivery_Category'] = df['Delivery_Category'].astype(str).str.upper()
df
```

```python
#Sorting
print(df.sort_values(by='Order_Value'))
df
```
